# Paso 4.2 Agregación de los Datos por Fecha, Tienda y Producto

```python
# 4. Agrupar por fecha, tienda y producto, sumando las cantidades y el monto de las ventas

ventas_diarias_agg = df_sales_full.groupby(['DateKey', 'StoreKey', 'ProductKey']).agg({
    'SalesAmount': 'sum',
    'SalesQuantity': 'sum'
}).reset_index()
```

```python
# Crear una columna de fecha basada en año y mes
dim_date['FullDate'] = dim_date['CalendarYear'].astype(str) + '-' + dim_date['MonthNumber'].astype(str)
```

# 4.3 Vamos a probar el modelado de ETC con Holt Winters usando Stastmodels

Ya que es bueno si hay tendencia y hay estacionalidad, ya que el hols linear trend es para serie scon tendencia pero sin estacionalidad y el suavizado simple sin tendencia ni estacionalidad

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# Paso 1: Ajustar el modelo Holt-Winters
# Este ejemplo usa el modelo multiplicativo para estacionalidad, ajuste según tus datos
model = ExponentialSmoothing(ventas_mensuales['SalesAmount'], 
                              trend='add',  # 'add' para tendencia aditiva, 'mul' para multiplicativa
                              seasonal='add',  # 'add' para estacionalidad aditiva, 'mul' para multiplicativa
                              seasonal_periods=12)  # Ajusta el período estacional según tu frecuencia (ej. 12 para mensual)
fit = model.fit()
```

```python

# Paso 2: Realizar pronósticos
forecast = fit.forecast(steps=12)  # Pronosticar los próximos 12 meses
```

```python
# Paso 3: Graficar resultados
plt.figure(figsize=(12, 6))
plt.plot(ventas_mensuales.index, ventas_mensuales['SalesAmount'], label='Datos Históricos')
plt.plot(forecast.index, forecast, label='Pronóstico', color='red')
plt.title('Pronóstico de Ventas Mensuales con Holt-Winters')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.legend()
plt.show()
```

```python
# Paso 4: Revisar el ajuste del modelo
print(fit.summary())
```
